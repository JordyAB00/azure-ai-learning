```python
def benchmark_modelo(model_name, task, prompt, max_length=50):
    """
    Evalúa un modelo midiendo tiempo y calidad de respuesta
    """
    print(f"\n{'='*60}")
    print(f"Modelo: {model_name}")
    print(f"Tarea: {task}")
    print(f"{'='*60}")

    # Cargar modelo
    print("Cargando modelo...")
    inicio_carga = time.time()

    try:
        pipe = pipeline(task, model=model_name)
        tiempo_carga = time.time() - inicio_carga
        print(f"✓ Cargado en {tiempo_carga:.2f} segundos")
    except Exception as e:
        print(f"✗ Error cargando: {e}")
        return None

    # Generar respuesta
    print(f"\nPrompt: \"{prompt}\"")
    print("Generando respuesta...")

    inicio_gen = time.time()
    try:
        resultado = pipe(prompt, max_length=max_length, do_sample=True)
        tiempo_gen = time.time() - inicio_gen

        if isinstance(resultado, list):
            texto_generado = resultado[0]['generated_text']
        else:
            texto_generado = resultado

        print(f"✓ Generado en {tiempo_gen:.2f} segundos\n")
        print(f"Respuesta:\n{texto_generado}")

        return {
            'modelo': model_name,
            'tiempo_carga': tiempo_carga,
            'tiempo_generacion': tiempo_gen,
            'respuesta': texto_generado
        }
    except Exception as e:
        print(f"✗ Error generando: {e}")
        return None
```

```python
# Prompt común para todos los modelos
prompt = "Los principales beneficios de la inteligencia artificial en empresas son"

# Lista de modelos a probar (todos small/medium para Colab)
modelos = [
    ("gpt2", "text-generation"),
    ("distilgpt2", "text-generation"),  # Versión comprimida de GPT-2
]

# Ejecutar benchmarks
resultados = []

for model_name, task in modelos:
    resultado = benchmark_modelo(model_name, task, prompt, max_length=80)
    if resultado:
        resultados.append(resultado)
    print("\n" + "="*60 + "\n")
```

```python
import pandas as pd

# Crear tabla comparativa
df = pd.DataFrame(resultados)

print("\n" + "="*60)
print("RESUMEN COMPARATIVO")
print("="*60 + "\n")

print(df[['modelo', 'tiempo_carga', 'tiempo_generacion']])

print("\n" + "="*60)
print("CONCLUSIONES")
print("="*60)

print(f"\n⚡ Modelo más rápido en carga: {df.loc[df['tiempo_carga'].idxmin(), 'modelo']}")
print(f"⚡ Modelo más rápido en generación: {df.loc[df['tiempo_generacion'].idxmin(), 'modelo']}")

print("\nNota: La calidad de respuestas debe evaluarse manualmente leyendo los textos generados.")
```
